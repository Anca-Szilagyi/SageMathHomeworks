```python
# (a)
reset()
popUS=[9.6,12.9,17.1,23.2,31.4,38.6,50.2,62.9,76,92,106.5,123.2]
popUS
```

```python
# Malthus model
t,r,x0,K,r0=var('t,r,x0,K,r0')
x=function('x')(t)
deqM=diff(x,t)==r*x
sM(t,x0,r)=desolve(deqM,[x,t],[0,x0])
sM(t,x0,r)
```

```python
deqV=diff(x,t)==r0*x*(1-x/K)
sV1=desolve(deqV,[x,t],[0,x0])
solve(sV1,x)
```

```python
eq1=sV1.log_simplify()
[sV2]=solve(eq1,x)
sV(t,x0,r0,K)=sV2.rhs()
sV(t,x0,r0,K)
```

```python
# (b)
plot([sM(t,10,0.1),sM(t,10,0.3),sM(t,2,0.5),sM(t,300,-0.1)],t,0,30,ymin=0,ymax=300)
```

```python
plot([sV(t,10,0.1,200),sV(t,10,0.2,200),sV(t,10,0.3,300),sV(t,200,-0.1,300)],t,0,50,ymin=0,ymax=300)
```

```python
assume(r,'real')
eq=sM(10,popUS[0],r)==popUS[1]
[r1]=solve(eq,r)
r2=r1.rhs()
r2
```

```python
r=numerical_approx(r2)
r
```

```python
assume(r0, 'real')
assume(K,'real')
eq1=sV(10,popUS[0],r0,K)==popUS[1]
eq2=sV(20,popUS[0],r0,K)==popUS[2]
s=solve([eq1,eq2],r0,K)
s
```

```python
eq22=eq2.substitute(r0==r00)
solve(eq22,K)
```
