```python
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)-diff(y,x)==1/(1+exp(x))
f(x,_K1,_K2)=desolve(eqd, y)
f(x,_K1,_K2)
plot(f(x, 1,2), x, -10, 10, ymin=-20, ymax=20)
```

```python
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)-5*diff(y,x)+4*y==0
f(x)=desolve(eqd, y, ics=[0, 5, 8])
f(x)
```

```python
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)-4*diff(y,x)+5*y==2*(x^2)*exp(x)
f(x)=desolve(eqd, y, ics=[0, 2, 3])
f(x)
```

```python
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x,2)+4*y==4*(sin(2*x)+cos(2*x))
f(x)=desolve(eqd, y, ics=[pi, 2*pi, 2*pi])
f(x)
```

```python
plot(f(x), x, -100, 100, ymin=-20, ymax=20)
```

```python
reset()
x,a=var('x, a')
y=function('y')(x)
eqd=diff(y,x,2)-diff(y, x)-2*y==0
f(x,a)=desolve(eqd, y, ivar=x, ics=[0, a, 2])
f(x,a)
```

```python
plot(f(x, -2), x, -100, 100, ymin=-20, ymax=20)
```
