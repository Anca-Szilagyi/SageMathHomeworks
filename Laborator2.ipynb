```python
reset()
x=var('x')
y=function('y')(x)
ecd=diff(y,x)==2*x*(1+y^2)
soli=desolve(ecd,y)
[sol]=solve(soli,y(x))
f(x,_C)=sol.rhs()
f(x,_C)
```

```python
plot(f(x,1/2), x, -pi/2, pi/2, ymin=-10, ymax=10)
```
