```python
# Exercitiul 2
# (a)
t = var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x+4*y
deq2=diff(y,t)==x+y
syst=[deq1,deq2]
in_cond=[0,1,2]
desolve_system(syst, [x,y], in_cond)
```

```python
sol=desolve_system(syst, [x,y], in_cond)
sol_x(t)=sol[0].rhs()
sol_y(t)=sol[1].rhs()
g1=plot(sol_x(t),t,-4,4,color='red',ymin=-20,ymax=20)
g2=plot(sol_y(t),t,-4,4,color='blue',ymin=-20,ymax=20)
show(g1+g2)
```
